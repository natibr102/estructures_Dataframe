#### Información que contienen las columnas:
* Año: 2008
* Mes :1-12
* Día del mes :1-31
* Día de la semana: 1 (lunes) - 7 (domingo)
* DepTime: hora de salida real (local, hhmm)
* CRSDepTime: hora de salida programada (local, hhmm)
* ArrTime: hora de llegada real (local, hhmm)
* CRSArrTime: hora de llegada programada (local, hhmm)
* UniqueCarrier: código de operador único
* FlightNum :número de vuelo
* Número de cola del avión TailNum: matrícula de la aeronave, identificador único de la aeronave
* Tiempo transcurrido real en minutos
* CRSElapsedTime: en minutos
* AirTime: en minutos
* Retraso en la llegada de ArrDelay, en minutos: un vuelo se cuenta como "a tiempo" si operó menos de 15 minutos después de la hora programada que se muestra en los Sistemas de Reservas Computarizados (CRS) de las aerolíneas.
* Retraso en la salida de DepDelay, en minutos
* Código de aeropuerto IATA de origen de origen
* Dest código de aeropuerto IATA de destino
* Distancia en millas
* TaxiEn taxi en tiempo, en minutos
* TaxiOut tiempo de salida del taxi en minutos
* Cancelado *fue cancelado el vuelo
* CancellationCode motivo de la cancelación (A = transportista, B = clima, C = NAS, D = seguridad)
* Desviado 1 = sí, 0 = no
* CarrierDelay en minutos: el retraso del transportista está bajo el control del transportista aéreo. Ejemplos de sucesos que pueden determinar el retraso del transportista son: limpieza de la aeronave, daños en la aeronave, espera de la llegada de pasajeros o tripulantes de conexión, equipaje, colisión con aves, carga de carga, catering, computadora, interrupción del equipo del transportista, legalidad de la tripulación (descanso del piloto o asistente) , daños por mercancías peligrosas, inspección de ingeniería, abastecimiento de combustible, manejo de pasajeros discapacitados, tripulación retrasada, servicio de lavabos, mantenimiento, sobreventas, servicio de agua potable, retiro de pasajeros ingobernables, embarque o asiento lentos, almacenamiento de equipaje de mano, retrasos por peso y equilibrio.
* WeatherDelay en minutos: el retraso del tiempo es causado por condiciones climáticas extremas o peligrosas que se pronostican o se manifiestan en el punto de partida, en ruta o en el punto de llegada.
* NASRetraso en minutos: el retraso que está bajo el control del Sistema Nacional del Espacio Aéreo (NAS) puede incluir: condiciones climáticas no extremas, operaciones aeroportuarias, volumen de tráfico pesado, control de tráfico aéreo, etc.
* Retraso de seguridad en minutos: el retraso de seguridad es causado por la evacuación de una terminal o vestíbulo, el reembarque de la aeronave debido a una violación de seguridad, equipo de control inoperativo y/o largas filas de más de 29 minutos en las áreas de control.
* LateAircraftDelay en minutos: Retraso en la llegada a un aeropuerto debido a la llegada tardía de la misma aeronave a un aeropuerto anterior. El efecto dominó de un retraso anterior en los aeropuertos posteriores se denomina propagación del retraso.

Fuente: https://www.kaggle.com/code/adveros/flight-delay-eda-exploratory-data-analysis?scriptVersionId=1821484&cellId=2

```python
 #Reducción del Dataset a 200.000 registros.
delay_short=delay.sample(n = 200000)
delay_short
```

#### Eliminación de columnas no relevantes: 
* Unnamed:no aporta ningún dato.

```python
new_delay=delay_short.drop (["Unnamed: 0",'DayOfWeek','Year','TailNum'], axis=1)
new_delay
```

#### Crea columnes noves (velocitat mitjana del vol, si ha arribat tard o no...).
* Nueva columna 'retraso', contiene el tiempo de demora entre la hora programada del vuelo y la hora real de salida, teniendo en cuenta que se considera retraso cuando la salida del vuelo supera los 15 minutos desde la hora programada de vuelo.


```python
new_delay['retraso'] = (new_delay['CRSDepTime'] + 15.0) - new_delay['DepTime']
new_delay['retraso']
```

* Nueva columna 'velocidad media de vuelo'.

```python
new_delay['velocidad media de vuelo']= new_delay['Distance']/(new_delay['AirTime']/60)
new_delay['velocidad media de vuelo']
```

* Nueva columna 'Arribo tarde', considerando que se considera retraso si arriba más de 15 minutos de la hora programada.

```python
import numpy as np
new_delay['Arribo tarde']=np.where(new_delay['ArrDelay']<15,0,1)
new_delay['Arribo tarde']
```
